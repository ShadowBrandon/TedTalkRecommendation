Objective: Have a pickle file of just Transcript, URL, index, tags, and title

```python
import pandas as pd
import pickle
```

```python
ted_main = pd.read_csv('./ted-talks/ted_main.csv')
ted_trans = pd.read_csv('./ted-talks/transcripts.csv')    

ted_all = pd.merge(ted_trans,right=ted_main,on='url')
ted_all.head(50)
```

```python
ted_all=ted_all.drop(columns=['comments','description','duration','event','film_date','languages','main_speaker','num_speaker','published_date','ratings','related_talks','speaker_occupation','views'])
ted_all['index']=ted_all.index
ted_all
```

```python
with open('ted_all_trimmed.pkl', 'wb') as picklefile:
    pickle.dump(ted_all, picklefile)
```
